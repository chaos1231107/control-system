```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('alt.txt', encoding='UTF-8')
df.columns = ['time', 'raw', 'sma']
df
```

```python
plt.plot(df['time'], df['raw'], label='Raw Data')
plt.plot(df['time'], df['sma'], label='SMA Data')
plt.xlabel('Time[ms]')
plt.ylabel('Altitude[m]')
plt.legend()
plt.show()
```
